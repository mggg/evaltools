```python
import matplotlib.pyplot as plt
from evaltools.plotting import (
    histogram,
    violin,
    arrow,
    ideal,
)
from evaltools.plotting.colors import districtr
import numpy as np
```

```python
scale = 1
scores = {
    "ensemble": [int(scale*x) for x in np.random.normal(14, 4, size=1000)],
    "citizen": [int(scale*x) for x in np.random.normal(12, 7, size=100)],
    "proposed": [scale*8, scale*9, scale*17],
}

proposed_info = {
    "colors": districtr(len(scores['proposed'])),
    "names": [f"Plan {i+1}" for i in range(len(scores['proposed']))],
}

fig, ax = plt.subplots(figsize=(12,6))
ax = histogram(ax, scores, proposed_info)
ax = arrow(ax, "Dem. advantage")
text = ax.set_xlabel("Split Counties", fontsize=24)
ax = ideal(ax, "Ideal Value", (10,16), "vertical", "red", alpha=0.2)
```

```python
means = [int(x) for x in np.random.normal(24, 6, size=10)]
labels = [f"E{i+1}" for i in range(len(means))]
num_proposed = 3

scores = {
    "ensemble": [[int(x) for x in np.random.normal(m, 4, size=1000)] for m in means],
    "proposed": [[int(x) for x in np.random.normal(m, 8, size=num_proposed)] for m in means],
}

proposed_info = {
    "colors": districtr(num_proposed),
    "names": [f"Plan {i+1}" for i in range(num_proposed)],
}
fig, ax = plt.subplots(figsize=(12,6))
ax = violin(ax, scores, labels, proposed_info, percentiles=(0,100))
arrow(ax, "Dem. seats", "vertical", padding=0.05)
text = ax.set_xlabel("Democratic Seats", fontsize=24)
ideal(ax, "Desired range", [20, 23], "horizontal")
```
